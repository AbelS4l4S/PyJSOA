```python
function_params = {
    #      f(x) | lb | ub | dim | global-minimun
    "F01": (Ackley,(-32,), (32,), 30, 0),
    "F02": (Brown,(-1,), (4,), 30, 0),
    "F03": (Exponential,(-1,), (1,), 30, -1),
    "F04": (Griewank,(-600,), (600,), 30, 0),
    "F05": (HappyCat,(-2,), (2,), 30, 0),
    "F06": (Michalewicz,(0,), (pi,), 30, -1.8013),
    "F08": (Periodic,(-10,), (10,), 30, 0.9),
    "F09": (Powell,(-1,), (1,), 30, 0),
    "F10": (Quartic,(-1.28,), (1.28,), 30, 0),
    "F11": (Rastrigin,(-5.12,), (5.12,), 30, 0),
    "F12": (Rosenbrock,(-5,), (10,), 30, 0),
    "F13": (RotatedHyperEllipsoid,(-65.536,), (65.536,), 30, 0),
    "F14": (Salomon,(-100,), (100,), 30, 0),
    "F15": (Schwefel2_20,(-100,), (100,), 30, 0),
    "F16": (Schwefel2_21,(-100,), (100,), 30, 0),
    "F17": (Schwefel2_22,(-100,), (100,), 30, 0),
    "F18": (Schwefel2_23,(-10,), (10,), 30, 0),
    "F19": (Sphere,(-5.12,), (5.12,), 30, 0),
    "F20": (SumSquares,(-10,), (10,), 30, 0),
    "F21": (Thevenot,(-2*pi,), (2*pi,), 30, -1),          # m=5, beta=15
    "F22": (XinSheYang,(-5,), (5,), 30, 0),
    "F23": (XinSheYangN2,(-2*pi,), (2*pi,), 30, 0),
    "F24": (XinSheYangN3,(-2*pi,), (2*pi,), 30, -1),      # m=5, beta=15
    "F25": (XinSheYangN4,(-10,), (10,), 30, -2)
    }


def initialization(pop_size, dim, ub, lb):

    if len(ub) == 1:
        vct_pop = np.random.uniform(lb, ub, (pop_size, dim))

    else:
        vct_pop = np.column_stack([np.random.uniform(lb[i], ub[i], pop_size) for i in range(len(ub))])
    return vct_pop


def get_algrthm_details(population, algorithm):

    j = np.random.choice(population.shape[0], size = 2, replace = False)
    best_vct = population[j[0], :]
    worst_vct = population[j[1], :]

    spyders_phrmns = np.zeros(population.shape[0])

    algrthm_details = {
        "jsoa" : [JSOA, [best_vct, worst_vct, spyders_phrmns]],
        "de"   : [DE_best_1_bin, [best_vct,]],
        "rnd-de": [DE_rand_1_bin, [best_vct,]],
        "de-2": [DE_best_2_bin, [best_vct,]],
        "rnd-de-2": [DE_rand_2_bin, [best_vct,]],
        "rnd-best-de":[DE_rand_to_best_1_bin, [best_vct,]],
        "rnd-best-de-2":[DE_rand_to_best_2_bin , [best_vct,],]
    }

    return algrthm_details.get(algorithm)
```

```python
def get_pheromone(fitness, min_val, v_max):
    pheromones = np.zeros_like(fitness)
    if min_val != v_max:
        pheromones = ((v_max - fitness) / (v_max - min_val))
    return pheromones


def cauchy_rand(m, c, rows):
    cauchy = c * np.tan(np.pi * (np.random.random(rows) - 0.5)) + m
    return cauchy


def JSOA(vcts_pop, generation, parameters, gravity = 9.80665, vo = 100):

    best_vct, worst_vct, spyders_phrmns = parameters
    recombination_rate = np.random.rand(vcts_pop.shape[0])

    attackers = np.where(recombination_rate < 0.5)[0]
    seekers = np.where(recombination_rate >= 0.5)[0]

    attack_rate = np.random.rand(attackers.size)
    search_rate = np.random.rand(seekers.size)

    jumpers = np.where(attack_rate < 0.5)[0]
    pursuers = np.where(attack_rate >= 0.5)[0]
    global_srchrs = np.where(search_rate < 0.5)[0]
    local_srchrs = np.where(search_rate >= 0.5)[0]

    # -----| 🏹 Brinca sobre la presa 🏹 |-----
    radians = 90 * random() * pi/180
    vcts_pop[attackers[jumpers]] = (vcts_pop[attackers[jumpers]] * tan(radians)) \
                                    - ((gravity * np.square(vcts_pop[attackers[jumpers]])) \
                                    / (2 * vo**2 * np.square(cos(radians))))

    # -----| 🏃‍♂️ Persigue la presa 🏃‍♂️ |-----
    j = np.random.choice(vcts_pop.shape[0], size=len(pursuers))
    j[pursuers == j] = np.random.choice(
                            np.setdiff1d(np.arange(0, vcts_pop.shape[0]), pursuers),
                            size = np.sum(pursuers == j))
    vcts_pop[attackers[pursuers]] = 0.5 * (vcts_pop[attackers[pursuers]] - vcts_pop[j])

    # -----| 🌎 Búsqueda global de la presa 🌎 |-----
    e1 = cauchy_rand(0, 1, global_srchrs.size)
    vcts_pop[seekers[global_srchrs]] = best_vct + (best_vct - worst_vct) * e1[:, np.newaxis]

    # -----| 🔎 Búsqueda local de la presa 🔎 |-----
    walk = -2 + 4 * np.random.rand(local_srchrs.size)
    e = np.random.normal(loc=0, scale=1, size=local_srchrs.size)
    vcts_pop[seekers[local_srchrs]] = best_vct + walk[:, np.newaxis] * (0.5 - e[:, np.newaxis])

    # -----| 🧪 Actualizaciòn de feromonas 🧪 |-----
    low_phrms = np.where(spyders_phrmns <= 0.3)[0]
    
    j = np.random.choice(vcts_pop.shape[0], size=(len(low_phrms), 2))
    j[:,1][j[:,0] == j[:,1]] = np.random.choice(
                            np.setdiff1d(np.arange(0, vcts_pop.shape[0]), j[:,0]),
                            size = np.sum(j[:,0] == j[:,1]))
    sigma = np.random.choice([-1,1], size=len(low_phrms))
    sigma = sigma[:, np.newaxis]
    vcts_pop[low_phrms] = best_vct + (vcts_pop[j[:,0]] - (sigma * vcts_pop[j[:,1]])) / 2
    
    return vcts_pop
```

```python
def shuffle_row(row):
    np.random.shuffle(row)
    return row


def generate_unique_random_matrix(m, num_columns, num_rows=None):
    num_rows = num_rows if num_rows is not None else m
    nums_list = np.arange(1, m)
    matrix = np.tile(nums_list, (num_rows, 1))

    matrix = np.apply_along_axis(shuffle_row, axis=1, arr=matrix)
    matrix = matrix[:, :num_columns]

    return matrix


def crossing(X,V,CR = 0.3):
    m, n = X.shape
    cross_rate = np.random.rand(m, n)
    
    rnbr = np.tile(range(1, n + 1), (m, 1))
    rnbr = np.apply_along_axis(shuffle_row, axis=1, arr=rnbr)
    
    j = np.arange(1, n+1)
    coincidencias = np.equal(rnbr, j)

    mask = (cross_rate <= CR) | coincidencias
    
    X[mask] = V[mask]

    return X

    
def DE_best_1_bin(population, generation, parameters):

    best_vct = parameters[0]
    mutated_pop = population.copy()
    f_values = np.random.uniform(-2, 2, size=population.shape[0])
    j = generate_unique_random_matrix(population.shape[0], 2)

    mutated_pop = best_vct + f_values[:, np.newaxis] * (population[j[:,0]] - population[j[:,1]])
    population = crossing(population, mutated_pop)

    return population


def DE_rand_1_bin(population, generation, parameters):

    best_vct = parameters[0]
    mutated_pop = population.copy()
    f_values = np.random.uniform(-2, 2, size=population.shape[0])
    j = generate_unique_random_matrix(population.shape[0], 3)

    mutated_pop = population[j[:,0]] + f_values[:, np.newaxis] * (population[j[:,1]] - population[j[:,2]])
    population = crossing(population, mutated_pop)

    return population


def DE_best_2_bin(population, generation, parameters):

    best_vct = parameters[0] 
    mutated_pop = population.copy()
    f_values = np.random.uniform(-2, 2, size=population.shape[0])
    j = generate_unique_random_matrix(population.shape[0], 4)

    mutated_pop = best_vct + f_values[:, np.newaxis] * (population[j[:,0]] + population[j[:,1]] - population[j[:,2]] - population[j[:,3]])
    population = crossing(population, mutated_pop)
    
    return population


def DE_rand_2_bin(population, generation, parameters):

    best_vct = parameters[0]
    mutated_pop = population.copy()
    f_values = np.random.uniform(-2, 2, size=population.shape[0])
    j = generate_unique_random_matrix(population.shape[0], 5)

    mutated_pop = population[j[:,0]] + f_values[:, np.newaxis] * (population[j[:,1]] + population[j[:,2]] - population[j[:,3]] - population[j[:,4]])
    population = crossing(population, mutated_pop)

    return population


def DE_rand_to_best_1_bin(population, generation, parameters):

    best_vct = parameters[0]
    mutated_pop = population.copy()
    f_values = np.random.uniform(-2, 2, size=population.shape[0])
    j = generate_unique_random_matrix(population.shape[0], 2)

    mutated_pop = population + f_values[:, np.newaxis] * (best_vct - population) + f_values[:, np.newaxis] * (population[j[:,0]]- population[j[:,1]])
    population = crossing(population, mutated_pop)

    return population


def DE_rand_to_best_2_bin(population, generation, parameters):

    best_vct = parameters[0]
    mutated_pop = population.copy()
    f_values = np.random.uniform(-2, 2, size=population.shape[0])
    j = generate_unique_random_matrix(population.shape[0], 4)

    mutated_pop = population + f_values[:, np.newaxis] * (best_vct - population) + f_values[:, np.newaxis] * (population[j[:,0]] - population[j[:,1]]) + f_values[:, np.newaxis] * (population[j[:,2]] - population[j[:,3]])
    population = crossing(population, mutated_pop)

    return population
```

```python
def optimize(pop_size, iterations, objctv_function, metahrstc):

    function, lb, ub, dim, optimal_value = function_params.get(objctv_function)
    population = initialization(pop_size, dim, ub, lb)
    convergence_graph = np.zeros([iterations])
    optimizer, params = get_algrthm_details(population, metahrstc)
    
    for i in range(iterations):

        trial_pop = population.copy()

        trial_pop = optimizer(trial_pop, i, params)
        trial_pop = np.clip(trial_pop,lb,ub)

        if i == 0:
            fitness = function(trial_pop)
            population = trial_pop.copy()

        else:
            new_fitness = function(trial_pop)
            bestfit_indx = np.where(new_fitness < fitness)[0]
            fitness[bestfit_indx] = new_fitness[bestfit_indx]
            population[bestfit_indx] = trial_pop[bestfit_indx]

        smallest_fitness = fitness.min()
        params[0] = population[fitness.argmin()]

        if metahrstc == "jsoa":
            greatest_fitness = fitness.max()
            params[1] = population[fitness.argmax()]
            params[2] = get_pheromone(fitness, smallest_fitness, greatest_fitness)

        convergence_graph[i] = smallest_fitness

    return convergence_graph, params[0], smallest_fitness
```

```python
def calculate_mae(iterations, test_functions, metahrstc):
    
    print(f"###### {metahrstc} ######")

    pop_size = 30
    generations = 200
    
    global_optimal = np.zeros([len(test_functions)])
    mean_optimal = np.zeros([len(test_functions)])
    std = np.zeros([len(test_functions)])
    
    for i, function in enumerate(test_functions):
        print(f"Función evaluándose: {function}")
        optimal_values = np.stack([optimize(pop_size, generations, function, metahrstc)[2] for _ in range(iterations)])
        
        f_min = function_params.get(function)[4]
        global_optimal[i] = np.min(optimal_values)
        mean_optimal[i] = np.mean(optimal_values)
        std[i] = np.std(optimal_values)
        
        print(f"Resultados las {iterations} iteraciones:")
        print(f"-- El optimo global encontrado es: {global_optimal[i]}")
        print(f"-- La media de los mejores {iterations} valores encontrados es: {mean_optimal[i]}")
        print(f"-- La desviación estandar del conjunto es: {std[i]}")
        print(f"-- El valor minimo de la función es: {f_min}")
        
        print("-------------------- \n")
    
    
    mean_abslt_err = np.sum(mean_optimal - global_optimal)/len(test_functions)

    return mean_abslt_err
```
